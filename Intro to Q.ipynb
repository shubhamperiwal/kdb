2) Cast 16:30:00 to int/float/minute/second <br/>
3) Cast the string “12:30:00” to time/minute/second <br/>
4) Cast 123.23 to real <br/>
5) Extract the month, week and year from the date 2015.06.03 <br/>
6) Get the current time/date/minute/second/hour <br/>
7) Cast 1234 to a string <br/>

```q
/1. 
type 234
type 234f
type 2010.05.10
type 12342342332 /int can't take such a big number. Use long
type "m"$2010.05 /convert to month using "$"
type type 500.05
type 3 1 2
type 3 1 2f
type (3f;1;2j)
```

```q
/ Cast 16:30:00 to int/float/minute/second 
"i"$16:30:00
"f"$16:30:00
"u"$16:30:00
`uu$16:30:00 /Get numerical minutes
`minute$16:30:00 /to minute precision
"v"$16:30:00
`ss$16:30:05 /Get numerical seconds
```

```q
/Cast the string “12:30:00” to time/minute/second 
"T"$"12:30:00" /capital letter so that it interprets whole string as a whole rather than each individual atom
"T"$("12:30:00"; "18:30:00")
"U"$"12:30:00"
`uu$"T"$"12:30:00"  /convert string to time and then to get minutes
"V"$"12:30:00"
```

```q
/Cast 123.23 to real 
"e"$123.23
```

```q
/Extract the month, week and year from the date 2015.06.03 
`mm$2015.06.03
`dd$2015.06.03

`week$2015.06.03
`year$2015.06.03
```

```q
/Get the current time/date/minute/second/hour 
`time$.z.T
.z.D
"u"$.z.T
`minute$.z.T
`mm$.z.T
"v"$.z.T
`second$.z.T
`ss$.z.T

`hh$.z.T
```
