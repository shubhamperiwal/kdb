```q
/ Cast 16:30:00 to int/float/minute/second 
"i"$16:30:00
"f"$16:30:00
"u"$16:30:00
`uu$16:30:00 /Get numerical minutes
`minute$16:30:00 /to minute precision
"v"$16:30:00
`ss$16:30:05 /Get numerical seconds
```

```q
/Cast the string “12:30:00” to time/minute/second 
"T"$"12:30:00" /capital letter so that it interprets whole string as a whole rather than each individual atom
"T"$("12:30:00"; "18:30:00")
"U"$"12:30:00"
`uu$"T"$"12:30:00"  /convert string to time and then to get minutes
"V"$"12:30:00"
```

```q
/Cast 123.23 to real 
"e"$123.23
```

```q
/Extract the month, week and year from the date 2015.06.03 
`mm$2015.06.03
`dd$2015.06.03

`week$2015.06.03
`year$2015.06.03
```

## Chapter 6: QSQL

#### Practice Question

1)	In a q session, create a table trade containing date, time, sym, side, price fields. Fill the table with 
10000 random records with the following constraints: <br>
>- 3 different dates (e.g. today, yesterday, day before yesterday) 
- random times (e.g. between 00:00:00.000 and now) 
- 5 different syms: \`VOD.L\`BMW.DE\`AAA.L\`FDP.L\`GOOG.NY 
- 2 sides:\`B\`S 
- price between 0 and 200 
<br>

a. Select from trade all buy side trades (\`B) made by \`AAA.L on one particular date. <br>
b. Repeat 2 applying the where constraints in the order side, sym, date. How long does it take to do this 1000 times? Repeat in the order date, sym, side. How long does this take? <br>
c. Select all trades made today with a price between 100 and 110 <br>
d. Generate a count of the number of trades each sym made with the same conditions as 4 <br>
e. On the third day, \`AAA.L was renamed to \`BBB.L. Update trade to reflect this change <br>
f. Delete all trades where the price was greater than 190 (keyword ‘delete’) <br>
g. Get a list of syms, that traded on the last 2 days, that end in .L (keyword ‘like’) <br>
h. By sym, calculate the maximum, minimum and average price as well as the spread (max-min) of trades made in the first minute of the second day on the buy side only. <br>

```q
table: ([]date:10000?.z.D-til 3; time: 10000?.z.T; sym: 10000?`VOD.L`BMW.DE`AAA.L`FDP.L`GOOG.NY;
    side: 10000?`B`S; price: 10000?200)
table
```

```q
//a. Select from trade all buy side trades (`B) made by `AAA.L on one particular date. 
select from table where sym=`AAA.L, side=`B, date=.z.D
```

```q
//b. Repeat 2 applying the where constraints in the order side, sym, date.
//How long does it take to do this 1000 times? Repeat in the order date, sym, side. How long does this take? 

select from table where side=`B, sym=`AAA.L, date=.z.D
```

```q
select date,time,sym,side,price from table where date=.z.D, sym=`AAA.L, side=`B
```

```q
//c. Select all trades made today with a price between 100 and 110
select from table where date=.z.D, price within (100;110)
```

```q
//d. Generate a count of the number of trades each sym made with the same conditions as 4 
select number: count i by sym from table
```

```q
//e. On the third day, `AAA.L was renamed to `BBB.L. Update trade to reflect this change
select sym from table where date=.z.D-2
```

```q
//f. Delete all trades where the price was greater than 190 (keyword ‘delete’) 
table_2: delete from table where price>190
//or
delete from `table where price>190
```

```q
//g. Get a list of syms, that traded on the last 2 days, that end in .L (keyword ‘like’) 
select distinct(sym) from table where date>.z.D-2, sym like "*.L"
```

```q
//h. By sym, calculate the maximum, minimum and average price as well as the spread (max-min) of trades made in the first 
//minute of the second day on the buy side only. 
select max_price:max(price), min_price:min(price), avg_price:avg(price), spread:max(price)-min(price) by sym from table where date=.z.D-1, ("u"$time)<00:01, side=`B
```
